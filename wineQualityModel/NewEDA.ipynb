# Outlier Detection and Removal on train set 

## Detection

```python
import numpy as np 
from scipy.stats import zscore

for col in X_train.columns:
    skew = X_train[col].skew()

    if abs(skew)>1:
        # IQR based Implementation
        Q1 = X_train[col].quantile(0.25)
        Q3 = X_train[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        higher_bound =Q3 + 1.5 * IQR
        X_train[col] = X_train[col].where(
            (X_train[col] >= lower_bound) & (X_train[col] <= higher_bound)
        )

    else:
        # Z-Score based Implementation
        z_scores = zscore(X_train[col])
        X_train[col] = X_train[col].where(np.abs(z_scores) <=3 )

```

# Feature Engineer

```python
X_train['density_ratio'] = X_train['residual sugar'] / X_train['density']
X_test['density_ratio'] = X_test['residual sugar'] / X_test['density']

X_train['sugar_acidity_ratio'] = X_train['residual sugar'] / (X_train['fixed acidity'] + 1e-5)
X_test['sugar_acidity_ratio'] = X_test['residual sugar'] / (X_test['fixed acidity'] + 1e-5)

X_train['free_to_total_sulfur_ratio'] = X_train['free sulfur dioxide'] / (X_train['total sulfur dioxide'] + 1e-5)
X_test['free_to_total_sulfur_ratio'] = X_test['free sulfur dioxide'] / (X_test['total sulfur dioxide'] + 1e-5)

X_train['citric_to_fixed_acidity'] = X_train['citric acid'] / (X_train['fixed acidity'] + 1e-5)
X_test['citric_to_fixed_acidity'] = X_test['citric acid'] / (X_test['fixed acidity'] + 1e-5)

X_train['acid_balance'] = X_train['fixed acidity'] / (X_train['pH'] + 1e-5)
X_test['acid_balance'] = X_test['fixed acidity'] / (X_test['pH'] + 1e-5)

X_train['sulfates_chlorides_ratio'] = X_train['sulphates'] / (X_train['chlorides'] + 1e-5)
X_test['sulfates_chlorides_ratio'] = X_test['sulphates'] / (X_test['chlorides'] + 1e-5)

X_train['alcohol_density_ratio'] = X_train['alcohol'] / X_train['density']
X_test['alcohol_density_ratio'] = X_test['alcohol'] / X_test['density']


X_train['alcohol_sulphates'] = X_train['alcohol'] * X_train['sulphates']
X_test['alcohol_sulphates'] = X_test['alcohol'] * X_test['sulphates']

X_train['volatile_acidity_sulphates'] = X_train['volatile acidity'] * X_train['sulphates']
X_test['volatile_acidity_sulphates'] = X_test['volatile acidity'] * X_test['sulphates']

```

# Preprocessing Pipelines

```python
# select num and cat 

num_cols = X_train.select_dtypes(include = 'number').columns
cat_cols = X_train.select_dtypes(include = 'object').columns
```

```python
# prepare pipeline for numerical column

from imblearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('scaler', StandardScaler() )
])
```

```python
# prepare pipeline for categorical 

from sklearn.preprocessing import OneHotEncoder

cat_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown = 'ignore'))
])
```

## combine pipeline

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer([
    ('num', num_pipe, num_cols),
    ('cat', cat_pipe, cat_cols),
])
```

```python
from sklearn.ensemble import RandomForestClassifier

model_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('sampler', sampler),
    ('classifier', RandomForestClassifier(random_state=42))
])
```
